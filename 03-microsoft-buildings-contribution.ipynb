```python
import pandas as pd

from src import Config
from src.persistence import create_duckdb_context, create_blob_storage_context
from src.utils import upload_parquet_to_container
```

```python
db = create_duckdb_context()
output_path = str(Config.OUTPUT_DATA_DIR / "microsoft.parquet")
```

```python
df = pd.read_csv("https://minedbuildings.z5.web.core.windows.net/global-buildings/dataset-links.csv", dtype=str)
df = df[df['Location'].str.lower() == 'norway']
dataset_paths = df["Url"].to_list()
```

```python
db.execute(
    f"""
    COPY (
        WITH ms_buildings AS (
            SELECT
                ST_GeomFromGeoJSON(geometry) AS geometry,
                NULLIF(properties.confidence, -1) AS confidence,
            FROM read_json_auto(?, format = 'newline_delimited')
        )

        SELECT md5(geometry) AS external_id, * FROM ms_buildings
    ) TO '{output_path}'
    """,
    [dataset_paths],
)
```

```python
storage_client = create_blob_storage_context()
upload_parquet_to_container(
    blob_service_client=storage_client,
    parquet_path=output_path,
    blob_name="microsoft.parquet",
    container_name="contributions"
)
```
